```python
#Q.3
from scipy.stats import norm
import numpy as np
  
data_start = -5
data_end = 5
data_points = 11
data = np.linspace(data_start, data_end, data_points)
  
mean = np.mean(data)
std = np.std(data)
  
probability_pdf = norm.pdf(3, loc=mean, scale=std)
print(probability_pdf)
```

```python
#Q.6
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
N = 500

data = np.random.randn(N)

count, bins_count = np.histogram(data, bins=10)
  
pdf = count / sum(count)
cdf = np.cumsum(pdf)

plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.legend()
```

```python
#Q.7
Binomial distribution describes the distribution of binary data from a finite sample. 
Thus it gives the probability of getting r events out of n trials. Poisson distribution describes the distribution of binary data from an infinite sample.
Thus it gives the probability of getting r events in a population.
```
